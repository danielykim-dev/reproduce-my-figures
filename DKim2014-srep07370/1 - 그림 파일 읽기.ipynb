아래와 같이 그림을 읽고 확인하는데 필요한 Library를 불러옵니다.

```python
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

그림 파일('crocus-3226433_640.jpg')은 아래 주소에서 내려받았습니다. 아래 그림은 Free License 입니다.

다른 파일을 내려받아 사용하셔도 됩니다.

이제 내려받은 그림 파일을 읽어봅시다.

아래와 같이 그림 파일을 읽을 수 있습니다.

```python
img = cv2.imread('crocus-3226433_640.jpg')
```

읽은 그림은 아래와 같이 확인할 수 있습니다.

```python
plt.imshow(img)
```

그림이 내려받기 전에 확인한 것과 약간 다르게 보입니다. 

그 이유는 그림 파일을 읽을 때 OpenCV에서 BGR 로 읽었기 때문입니다.

아래와 같이 그림의 채널을 BGR 을 RGB로 바꿔서 내려받은 페이지에서 확인한 것과 같게 만들 수 있습니다.

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

이제 내려받은 페이지에서 봤던 것과 같게 보입니다.

```python
plt.imshow(img_rgb)
```
